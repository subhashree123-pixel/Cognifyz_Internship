```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dataset .csv")
```

```python
df.head()
```

```python
df["Price range"].value_counts()
```

```python
price_range_count = df["Price range"].value_counts()

print(price_range_count)
```

```python
average_rating = df.groupby("Price range")["Aggregate rating"].mean()

print(average_rating)
```

```python
highest_rating = average_rating.idxmax()

print("Price Range with Highest Average Rating:", highest_rating)
```

```python
print("Highest Average Rating:", average_rating.max())
```

```python
price_range_count.plot(kind="bar")

plt.title("Number of Restaurants by Price Range")
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")

plt.show()
```

```python
average_rating.plot(kind="bar")

plt.title("Average Rating by Price Range")
plt.xlabel("Price Range")
plt.ylabel("Average Rating")

plt.show()
```
