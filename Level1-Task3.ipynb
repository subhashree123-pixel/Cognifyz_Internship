```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dataset .csv")
```

```python
df["Price range"].head(5)
                      
```

```python
price_count = df["Price range"].value_counts().sort_index()
print(price_count)
```

```python
total_restaurants = len(df)
price_percentage = (price_count / total_restaurants)*100
print(price_percentage)
```

```python
plt.figure(figsize=(8,5), facecolor="#2E2E2E")

ax = plt.gca()
ax.set_facecolor("#2E2E2E")

colors = ["#4CAF50", "#2196F3", "#FFC107", "#F44336"]
plt.grid(axis='y', linestyle='--', alpha=0.3)

for i, value in enumerate(price_count):
    plt.text(i, value + 50, str(value), ha='center', color='white')

price_count.plot(kind = "bar", color=colors)

plt.title("Distribution of Restaurants by Price Range",color = "white",fontsize=14)
plt.ylabel("Number of Restaurants",color="white")
plt.xlabel("Price Range", color="white")

plt.xticks(rotation=0,color="white")
plt.yticks(rotation=0,color="white")

plt.show()
```
