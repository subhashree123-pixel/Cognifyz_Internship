City Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("Dataset .csv")   # or "data.csv" if you renamed it
```

```python
df
```

```python
df["City"].head()
```

```python
city_count = df["City"].value_counts()
print(city_count)
```

```python
top_city = city_count.head(5)
print(top_city)
```

```python
plt.figure(figsize=(8,5), facecolor = "black")

ax = plt.gca()
ax.set_facecolor("black")

colors = ["blue", "green", "orange", "red", "yellow"]

top_city.plot(kind = "bar", color=colors, ax = ax)

plt.title("City with Highest Number of Restaurant", color="white",fontsize=14)
plt.xlabel("City", color="white")
plt.ylabel("Numbers of Restaurants", color="white")

plt.xticks(color="white")
plt.yticks(color="white")

plt.show()
```

```python
df.columns
```

```python
city_rating = df.groupby("City")["Aggregate rating"].mean()
print(city_rating)
```

```python
highest_rating = city_rating.sort_values(ascending=False).head(10)
print(highest_rating)
```

```python
plt.figure(figsize=(10,6), facecolor = "gray")

ax = plt.gca()
ax.set_facecolor("gray")

colors = ["yellow", "maroon", "skyblue", "pink", "white", "purple", "orange", "green", "lime", "red"]

highest_rating.plot(kind="bar", color=colors, ax=ax)

plt.title("Top 10 Cities by Average Rating", color="black",fontsize=14)
plt.xlabel("Top 10 Cities", color="black")
plt.ylabel("Average Rating", color="black")

plt.xticks(color="black")
plt.yticks(color="black")

plt.show()
```
