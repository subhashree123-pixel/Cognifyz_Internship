```python
import pandas as pd
```

```python
df = pd.read_csv("Dataset .csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df["Restaurant Name Length"] = df["Restaurant Name"].str.len()
```

```python
df[["Restaurant Name", "Restaurant Name Length"]].head()
```

```python
df["Address Length"] = df["Address"].str.len()
```

```python
df[["Address", "Address Length"]].head()
```

```python
df["Table Booking"] = df["Has Table booking"].map({
    "Yes":1,
    "No":0
})
```

```python
df[["Has Table booking","Table Booking"]].head()
```

```python
df["Online Delivery"] = df["Has Online delivery"].map({
    "Yes":1,
    "No":0
})
```

```python
df[["Has Online delivery","Online Delivery"]].head()
```

```python
df.head()
```

```python
df.to_csv("Feature_Engineered_Dataset.csv", index=False)
```

```python
df["Popularity"] = pd.cut(
    df["Votes"],
    bins=[0, 100, 500, df["Votes"].max()],
    labels=["Low", "Medium", "High"]
)

df[["Votes", "Popularity"]].head()
```
