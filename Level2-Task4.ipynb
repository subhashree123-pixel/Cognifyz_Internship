```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dataset .csv")
```

```python
df.head()
```

```python
df[["Latitude", "Longitude"]].head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6), facecolor='black')

ax = plt.gca()
ax.set_facecolor('black')

plt.scatter(
    df["Longitude"],
    df["Latitude"],
    color='lime',     
    alpha=0.7
)

plt.title("Restaurant Locations", color='white', fontsize=14)
plt.xlabel("Longitude", color='white')
plt.ylabel("Latitude", color='white')

# Change axis tick colors
plt.xticks(color='white')
plt.yticks(color='white')

# Change axis border colors
for spine in ax.spines.values():
    spine.set_color('white')

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6), facecolor='black')

ax = plt.gca()
ax.set_facecolor('black')

scatter = plt.scatter(
    df["Longitude"],
    df["Latitude"],
    c=df["Aggregate rating"],   
    cmap="plasma",             
    alpha=0.8
)

# Color bar
cbar = plt.colorbar(scatter)
cbar.set_label("Aggregate Rating", color='white')
cbar.ax.yaxis.set_tick_params(color='white')
plt.setp(cbar.ax.get_yticklabels(), color='white')

# Make the color bar border visible
cbar.outline.set_edgecolor('white')

# Title and labels
plt.title("Restaurant Ratings by Location", color='white', fontsize=14)
plt.xlabel("Longitude", color='white')
plt.ylabel("Latitude", color='white')

# Axis tick colors
plt.xticks(color='white')
plt.yticks(color='white')

# Axis border colors
for spine in ax.spines.values():
    spine.set_color('white')

plt.show()     
```

```python
city_count = df["City"].value_counts()

print(city_count)
```

```python
top_cities = df["City"].value_counts().head(10)

print(top_cities)
```

```python
top_cities.plot(kind="bar", figsize=(10,5))

plt.title("Top 10 Cities by Number of Restaurants")
plt.xlabel("City")
plt.ylabel("Number of Restaurants")

plt.show()
```

```python
city_rating = df.groupby("City")["Aggregate rating"].mean()

city_rating.head()
```

```python
top_rating = city_rating.sort_values(ascending=False).head(10)

print(top_rating)
```

```python
top_rating.plot(kind="bar", figsize=(10,5))

plt.title("Top 10 Cities by Average Rating")
plt.xlabel("City")
plt.ylabel("Average Rating")

plt.show()
```
