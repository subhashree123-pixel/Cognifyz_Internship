```python
df = df.dropna(subset=["Cuisines"])
```

```python
df
```

```python
cuisine_count = df["Cuisines"].value_counts()
```

```python
print(cuisine_count)
```

```python
cuisine_count = df["Cuisines"].str.split(", ").explode().value_counts()

top3 = cuisine_count.head(3)
print(top3)
```

```python
top3 = cuisine_count.head(3)
print(top3)
```

```python
total_restaurants = len(df)
print(total_restaurants)
```

```python
percentage = (top3/total_restaurants)*100
```

```python
print(percentage)
```

```python
plt.figure(figsize=(8,5), facecolor = "black")

ax = plt.gca()
ax.set_facecolor("black")

colors = ["blue", "green", "orange"]

top3.plot(kind = "bar", color=colors, ax = ax)

plt.title("Top 3 Most Common Cuisines", color="white",fontsize=14)
plt.xlabel("Cuisines", color="white")
plt.ylabel("Numbers of Restaurants", color="white")

plt.xticks(color="white")
plt.yticks(color="white")

plt.show()
```
