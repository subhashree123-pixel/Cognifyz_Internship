```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dataset .csv")
```

```python
df
```

```python
df.columns
```

```python
delivery_count = df["Has Online delivery"].value_counts()
print(delivery_count)
```

```python
total_restaurants = len(df)
delivery_percentage = (delivery_count/total_restaurants)*100
print(delivery_percentage)
```

```python
rating = df.groupby("Has Online delivery")["Aggregate rating"].mean()
print(rating)
```

```python
plt.figure(figsize=(9,5), facecolor="gray")

ax = plt.gca()
ax.set_facecolor("black")

colors = ["#FF5252", "#00E676"]

delivery_count.plot(kind="bar", color=colors, ax=ax)

plt.title("Restaurants Offering Online Delivery", color="white", fontsize=14)
plt.xlabel("Online Delivery", color="white")
plt.ylabel("Number of Restaurants", color="white")

plt.xticks(rotation=0, color="white")
plt.yticks(color="white")

plt.show()
```
