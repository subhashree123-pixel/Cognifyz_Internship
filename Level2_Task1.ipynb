```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dataset .csv")
```

```python
df
```

```python
df.head(5)
```

```python
df.columns
```

```python
df["Has Table booking"].value_counts()
```

```python
table_booking_count = df["Has Table booking"].value_counts()

print(table_booking_count)
```

```python
table_booking_percentage = (
    df["Has Table booking"].value_counts(normalize=True) * 100
)

print(table_booking_percentage)
```

```python
df["Has Online delivery"].value_counts()
```

```python
online_delivery_count = df["Has Online delivery"].value_counts()

print(online_delivery_count)
```

```python
online_delivery_percentage = (
    df["Has Online delivery"].value_counts(normalize=True) * 100
)

print(online_delivery_percentage)
```

```python
df.groupby("Has Table booking")["Aggregate rating"].mean()
```

```python
table_rating = (
    df.groupby("Has Table booking")["Aggregate rating"]
      .mean()
)

print(table_rating)
```

```python
online_rating = (
    df.groupby("Has Online delivery")["Aggregate rating"]
      .mean()
)

print(online_rating)
```

```python
table_booking_count.plot(kind="bar")

plt.title("Restaurants with Table Booking")
plt.xlabel("Table Booking")
plt.ylabel("Number of Restaurants")

plt.show()
```

```python
online_delivery_count.plot(kind="bar")

plt.title("Restaurants with Online Delivery")
plt.xlabel("Online Delivery")
plt.ylabel("Number of Restaurants")

plt.show()
```

```python
table_rating.plot(kind="bar")

plt.title("Average Rating by Table Booking")
plt.xlabel("Table Booking")
plt.ylabel("Average Rating")

plt.show()
```

```python
online_rating.plot(kind="bar")

plt.title("Average Rating by Online Delivery")
plt.xlabel("Online Delivery")
plt.ylabel("Average Rating")

plt.show()
```
